### Puxar base de dados

```python
import pandas as pd

exportacao_franca = pd.read_csv("exportacoes_franca.csv")
exportacao_franca = exportacao_franca.loc[exportacao_franca["Economic Block"] == "Europe", :]
display(exportacao_franca)
```

### Informações Gerais

```python
print(exportacao_franca.info())
```

### Como foi a evolução das exportações para a frança ao longo dos anos?

```python
exportacao_ano = exportacao_franca[["Year", "US$ FOB"]]
exportacao_ano = exportacao_ano.groupby("Year").sum()
exportacao_ano.plot(kind="bar")

# função de formatação
def formatar(valor):
    return f"${valor:,.2f}"

# aplicar função
exportacao_ano["US$ FOB"] = exportacao_ano["US$ FOB"].map(formatar)

display(exportacao_ano)
```

### Quais os produtos mais exportados ao longo de todo o período?

```python
exportacao_produto = exportacao_franca[["SH4 Description", "US$ FOB"]]
exportacao_produto = exportacao_produto.groupby("SH4 Description").sum()
exportacao_produto = exportacao_produto.sort_values("US$ FOB", ascending=False)
exportacao_produto["US$ FOB"] = exportacao_produto["US$ FOB"].map(formatar)
display(exportacao_produto)
```

### Em 2020 qual cidade mais exportou para a França?

```python
exportacao_2020 = exportacao_franca.loc[exportacao_franca["Year"] == 2020, :]
exportacao_cidade_2020 = exportacao_2020[["City", "US$ FOB"]]
exportacao_cidade_2020 = exportacao_cidade_2020.groupby("City").sum()
exportacao_cidade_2020 = exportacao_cidade_2020.sort_values("US$ FOB", ascending=False)
exportacao_cidade_2020["US$ FOB"] = exportacao_cidade_2020["US$ FOB"].map(formatar)
display(exportacao_cidade_2020) 
```

### O que as 2 maiores cidades exportaram tanto?

```python
cidade = exportacao_cidade_2020.index[0]

tabela_cidade = exportacao_2020.loc[exportacao_2020["City"] == cidade, :]
tabela_cidade = tabela_cidade.groupby("SH4 Description").sum()
tabela_cidade = tabela_cidade[["US$ FOB"]]
tabela_cidade = tabela_cidade.sort_values("US$ FOB", ascending=False)
tabela_cidade["US$ FOB"] = tabela_cidade["US$ FOB"].map(formatar)
display(tabela_cidade)
```
